```python
# import pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Month': ['Sep 2025', 'Oct 2025', 'Nov 2025', 'Dec 2025', 'Jan 2026', 'Feb 2026', 'Mar 2026', 'Apr 2026', 'May 2026', 'June 2026', 'July 2026'],
    'Flight Rate Per Person': ['133-337', '133-337', '133-337', '133-337', '133-337', '133-337', '167-337', '167-337', '167-337', '167-337', '167-337'],
    'Average Flight Cost Per Person': [235, 235, 235, 235, 235, 235, 252, 252, 252, 252, 252],
    'Average Flight Cost For Family of Five': [1175, 1175, 1175, 1175, 1175, 1175, 1260, 1260, 1260, 1260, 1260],
    'Average Hotel Cost Per Night': [210, 230, 240, 290, 220, 230, 270, 280, 250, 300, 310],
    'Car Rental Fees For 5 Days': [125, 150, 150, 175, 125, 125, 150, 150, 175, 175, 175], 
    'Average Ticket Cost  ': [130, 140, 145, 160, 135, 140, 150, 155, 145, 160, 165]

}
df_disney = pd.DataFrame(data)
```

```python
# load in the data frame 
df_disney
```

```python
# add in the peak visitation rate 
df_disney['Estimated Peak Visitation Rate'] = ['Low (School Resumes)', 'Moderate (Halloween Events)', 'High (Thanksgiving)', 'Very High (Holiday Season)','Low (Post-holiday dip)', 'Moderate', 'High (Spring Break)', 'High (Easter/Spring Break)', 'Moderate', 'Very High (Summer starts)', 'Very High (Peak summer)']
```

```python
# check addition
df_disney
```

```python
# add a column for hotel 5 hotel nights 
df_disney['Average Hotel Cost for 5 Days'] = df_disney['Average Hotel Cost Per Night'] * 5
```

```python
# check updated data 
df_disney
```

```python
# check to ensure all columns are up to date 
print(df_disney.columns.tolist())
```

```python
# remove any unnecessary spaces from the table columns 
df_disney.columns = df_disney.columns.str.strip()
```

```python
# add a total cost column to the table 
# if this fails just print the list of columns in the df and see if the format is the same
for item in ['Average Flight Cost For Family of Five', 'Average Hotel Cost for 5 Days', 'Car Rental Fees For 5 Days', 'Average Ticket Cost']:
    df_disney[item] = pd.to_numeric(df_disney[item], errors = 'coerce')

df_disney.fillna(0, inplace=True)
df_disney['Total Cost'] = df_disney['Average Flight Cost For Family of Five'] + df_disney['Average Hotel Cost for 5 Days'] + df_disney['Car Rental Fees For 5 Days'] + df_disney['Average Ticket Cost']
```

```python
# check updated df 
df_disney
```

Data Analysis Plots 

1. Scatter Plot (Compare Month to Total Cost)

```python
plt.figure(figsize=(12,12))
plt.scatter(df_disney['Month'], df_disney['Total Cost'], c='blue', s=50, marker='*')
plt.title('Estimated Total Disney Trip Cost Each Month', fontsize = 14)
plt.xlabel('Month', fontsize = 12)
plt.ylabel('Total Cost (USD)', fontsize = 12)
plt.show()
```

2. Distribution of Total Cost Breakdown

```python
labels = ['Flight', 'Ticket', 'Rental Car', 'Hotel']
row = df_disney[df_disney['Month'] == 'Sep 2025'].iloc[0]
sizes = [
    row['Average Flight Cost For Family of Five'], 
    row['Average Ticket Cost'],
    row['Car Rental Fees For 5 Days'],
    row['Average Hotel Cost for 5 Days']
]

# create pie chart 
plt.figure(figsize=(8,8))
plt.pie(sizes, labels=labels)

# update the title for different months
plt.title('Disney Trip Cost Breakdown for September', fontsize = 12)
plt.axis('equal') #ensures circular pie chart 
plt.tight_layout()
plt.show

```

Cost Variation Across Months 

```python
plt.figure(figsize=(12,12))
plt.plot(df_disney['Month'], df_disney['Average Ticket Cost'])
plt.title('Average Disney Ticket Cost Per Month', fontsize = 15)
plt.xlabel('Month', fontsize = 15)
plt.ylabel('Disney Ticket Cost (USD)', fontsize = 15)
plt.show()
```
